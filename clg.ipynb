```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=['mon','tues','wed','thurs','fri']
y=[7,8,9,10,11]
ax=sns.stripplot(x,y)
plt.show()
```

```python
sns.set(style='whitegrid')
iris=sns.load_dataset('iris')
ax=sns.stripplot(data=iris)
plt.show()
```
